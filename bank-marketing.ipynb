```python
df=pd.read_csv("/kaggle/input/datasets/janiobachmann/bank-marketing-dataset/bank.csv") 
#Read data file
df
```

```python
df=df.drop(columns =['job'])
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
#leColumns=['marital','education','default','housing','loan','contact','poutcome','deposit']
df['marital']=le.fit_transform(df['marital'])
df['education']=le.fit_transform(df['education'])
df['default']=le.fit_transform(df['default'])
df['housing']=le.fit_transform(df['housing'])
df['loan']=le.fit_transform(df['loan'])
df['contact']=le.fit_transform(df['contact'])
df['poutcome']=le.fit_transform(df['poutcome'])
df['deposit']=le.fit_transform(df['deposit'])
df['month']=le.fit_transform(df['month'])
#df=df.drop(columns='month')
df
```

```python
X=df.drop(columns='deposit')
y=df['deposit']
from sklearn.feature_selection import GenericUnivariateSelect
from sklearn.feature_selection import f_classif
FeatureSelect=GenericUnivariateSelect(score_func=f_classif ,param=65)
X_fclassif=FeatureSelect.fit_transform(X,y)
#X_fclassif.shape
#feature selection using GenericUnivariateSelect (f_classif0) 
FeatureSelect.get_support()
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(random_state=42)
Feature_select=SelectFromModel(model ,threshold=-np.inf,max_features=10)
#threshold=-np.inf to enable max features =10
X_SFM=Feature_select.fit_transform(X,y)
X_SFM.shape
#feature selection using SelectFromModel (RandomForestClassifier) 
Feature_select.get_support()
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train,y_test=train_test_split(
    X_fclassif,
    y,
    test_size=0.25,
    random_state=44,
    shuffle=True  
) #data split (test,train) fclassif
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()
model.fit(X_train,y_train)
fclassif_predict=model.predict(X_test)
#train on fclassif selection features
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, fclassif_predict)
print(accuracy*100)
#fclassif acuuracy predict =81%
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, fclassif_predict))
```

```python
#Select from model 
#split
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train,y_test=train_test_split(
    X_SFM,
    y,
    test_size=0.25,
    random_state=44,
    shuffle=True  
)
#train
model = GradientBoostingClassifier()
model.fit(X_train,y_train)
SFM_predict=model.predict(X_test)
#train on SFM featurs
```

```python
accuracy = accuracy_score(y_test, SFM_predict)
print(accuracy*100)
#SFM features accuracy =83%
```

```python
print(classification_report(y_test, SFM_predict))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, SFM_predict))
```

```python
#randomfort model
model=RandomForestClassifier()
model.fit(X_train,y_train)
RF_predict=model.predict(X_test)
accuracy = accuracy_score(y_test, RF_predict)
print(accuracy*100)
```

```python
print(classification_report(y_test, RF_predict))
```

```python
print(confusion_matrix(y_test, RF_predict))
# Random Frost achieved better performance in terms of accuracy and recall.
# It also reduced false negatives, meaning fewer missed potential customers.
```
